## 3.1) Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pie chart for distribution of feature
def plot_pie_chart(df, feature, figsize=(6, 6), color_map='pastel', pctdistance=0.7, explode_factor=0.02):
    # Count the occurrences of each class in the feature
    target_counts = df[feature].value_counts()
    
     # Creating a pie chart
    plt.figure(figsize=figsize)
    colors = sns.color_palette(color_map)[0:len(target_counts)]
    explode = [explode_factor] * len(target_counts)  # Explode each slice by the specified factor
    plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', colors=colors, startangle=140, explode=explode, pctdistance=pctdistance)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# Pie plot of Transported
plot_pie_chart(training_data, 'Transported')
```

```python
# Pie plot of Transported
plot_pie_chart(training_data, 'CryoSleep', color_map='muted')
```

```python
# Function to analyze categorical variable
def plot_bar_chart(df, feature, all=True):
    # Counting the occurrences of each category
    feature_counts = df[feature].value_counts()

    if all:
        pass
    else:
        feature_counts = feature_counts.head(10)
        
    feature_percentages = (feature_counts / len(df)) * 100

    print(f"Counts for {feature}:")
    print(feature_counts, "\n")
    print(f"Percentages for {feature}:")
    print(feature_percentages, "\n")
    
    # Creating a bar plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x=feature_counts.index, y=feature_counts.values, palette='pastel')
    
    # Adding percentages on top of the bars
    for index, value in enumerate(feature_counts.values):
        plt.text(index, value, f'{value} ({feature_percentages.values[index]:.1f}%)', ha='center', va='bottom')
    
    plt.title(f'Bar Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

## 3.2) Bivariate Analysis

```python
def transportation_rate_by_category(df, categorical_feature, figsize=(10, 6)):
    target_feature='Transported'
    
    # Calculate transportation rate for each category
    category_counts = df.groupby(categorical_feature)[target_feature].mean() * 100
    
    # Print transportation rates
    print(f"Transportation rates for {categorical_feature}:")
    print(category_counts, "\n")
    
    # Create a bar plot
    plt.figure(figsize=figsize)
    sns.barplot(x=category_counts.index, y=category_counts.values)
    
    # Add percentages on top of the bars
    for index, value in enumerate(category_counts.values):
        plt.text(index, value, f'{value:.1f}%', ha='center', va='bottom')
    
    plt.title(f'Transportation Rate by {categorical_feature}')
    plt.xlabel(categorical_feature)
    plt.ylabel('Transportation Rate (%)')
    plt.show()
```

```python
# Analyzing transportation rate by HomePlanet
transportation_rate_by_category(training_data, 'HomePlanet')
```

```python
# Analyzing transportation rate by CryoSleep
transportation_rate_by_category(training_data, 'CryoSleep', (5, 4))
```

```python
# Analyzing transportation rate by Destination
transportation_rate_by_category(training_data, 'Destination')
```

```python
# Analyzing transportation rate by VIP
transportation_rate_by_category(training_data, 'VIP', (6, 4))
```

```python
def transportation_rate_by_numerical_feature(df, numerical_feature, bins=10):
    target_feature='Transported'
    
    # Create bins for the numerical feature
    df['binned'] = pd.cut(df[numerical_feature], bins=bins)
    
    # Calculate transportation rate for each bin
    bin_transportation_rate = df.groupby('binned')[target_feature].mean() * 100
    
    # Print transportation rates
    print(f"Transportation rates for {numerical_feature} (binned):")
    print(bin_transportation_rate, "\n")
    
    # Create a bar plot
    plt.figure(figsize=(12, 6))
    sns.barplot(x=bin_transportation_rate.index.astype(str), y=bin_transportation_rate.values, palette='pastel')
    
    # Add percentages on top of the bars
    for index, value in enumerate(bin_transportation_rate.values):
        plt.text(index, value, f'{value:.1f}%', ha='center', va='bottom')
    
    plt.title(f'Transportation Rate by {numerical_feature} (binned)')
    plt.xlabel(f'{numerical_feature} (binned)')
    plt.ylabel('Transportation Rate (%)')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Transportation rate by RoomService
transportation_rate_by_numerical_feature(training_data, 'RoomService', 5)
```

```python
# Transportation rate by Age
transportation_rate_by_numerical_feature(training_data, 'Age', 5)
```

```python
# Transportation rate by FoodCourt
transportation_rate_by_numerical_feature(training_data, 'FoodCourt', 5)

# Transportation rate by ShoppingMall
transportation_rate_by_numerical_feature(training_data, 'ShoppingMall', 5)

# Transportation rate by Spa
transportation_rate_by_numerical_feature(training_data, 'Spa', 5)

# Transportation rate by VRDeck
transportation_rate_by_numerical_feature(training_data, 'VRDeck', 5)
```

# 4) Data Preprocessing

## a) Training data Preparation

```python
# Dropping cabin as it has a lot of values
training_data.drop("Cabin", axis=1, inplace=True)
```

```python
# Checking null values
training_data.isnull().sum()
```

```python
# Checking duplicated records
training_data[training_data.duplicated()]
```

I fill missing numerical values with the median and categorical with the most frequent value

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Columns by type
numeric_features = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Age']
categorical_features = ['HomePlanet', 'CryoSleep', 'Destination', 'VIP']
```

```python
# Pipeline for imputing null in numeric and Feature Scaling
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Encoding Categorical Features and Scaling Numerical Features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', Pipeline(steps=[
            ('impute', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ]), categorical_features)
    ])
```

```python
# Preprocessing the data
X = training_data.drop(columns=['Transported', 'PassengerId', 'Name'])
y = training_data['Transported']

# Applying transformations to the training data
X_preprocessed = preprocessor.fit_transform(X)
```

```python
# Converting the preprocessed data back to a DataFrame
preprocessed_columns = numeric_features + list(preprocessor.named_transformers_['cat']['onehot'].get_feature_names_out(categorical_features))
X_preprocessed_df = pd.DataFrame(X_preprocessed, columns=preprocessed_columns)
X_preprocessed_df.head()
```

```python
print(f"Shape of the preprocessed dataset: {X_preprocessed.shape}")
```

```python
# Splitting data into train and test
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X_preprocessed, y, test_size=0.2, random_state=42)

# Print the shapes of the resulting datasets
print(f"Shape of X_train: {X_train.shape}")
print(f"Shape of X_test: {X_test.shape}")
print(f"Shape of y_train: {y_train.shape}")
print(f"Shape of y_test: {y_test.shape}")
```

## b) Testing Data Preparation

```python
# Dropping irrelavant columns
X_test_data = testing_data.drop(columns=['PassengerId', 'Name', "Cabin"])

# Preprocessing data
X_test_preprocessed = preprocessor.transform(X_test_data)
```

```python
# Printing the shape of the preprocessed testing data
print(f"Shape of the preprocessed testing data: {X_test_preprocessed.shape}")

# Dataframe
preprocessed_columns = numeric_features + list(preprocessor.named_transformers_['cat']['onehot'].get_feature_names_out(categorical_features))
X_test_preprocessed_df = pd.DataFrame(X_test_preprocessed, columns=preprocessed_columns)

print("Preprocessed Testing Data:")
X_test_preprocessed_df.head()
```

# 4) ML Modelling

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
from sklearn.metrics import RocCurveDisplay

def comprehensive_evaluation(y_test, y_pred, class_names, y_proba=None):
    # Calculate metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, average='weighted')
    recall = recall_score(y_test, y_pred, average='weighted')
    f1 = f1_score(y_test, y_pred, average='weighted')
    
    # Print evaluation metrics
    print(f"Accuracy: {accuracy:.4f}")
    print(f"Precision: {precision:.4f}")
    print(f"Recall: {recall:.4f}")
    print(f"F1 Score: {f1:.4f}")
    
    # Print classification report
    print("\nClassification Report:")
    print(classification_report(y_test, y_pred, target_names=class_names))
    
    # Generate and plot confusion matrix
    conf_matrix = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(8, 6))
    disp = ConfusionMatrixDisplay(confusion_matrix=conf_matrix, display_labels=class_names)
    disp.plot(cmap=plt.cm.Blues)
    plt.title("Confusion Matrix")
    plt.show()
    
    # Plot ROC AUC curve if probability scores are provided
    if y_proba is not None:
        fpr, tpr, _ = roc_curve(y_test, y_proba[:, 1])
        roc_auc = roc_auc_score(y_test, y_proba[:, 1])
        plt.figure(figsize=(8, 6))
        RocCurveDisplay(fpr=fpr, tpr=tpr, roc_auc=roc_auc).plot()
        plt.title(f"ROC AUC Curve (AUC = {roc_auc:.4f})")
        plt.show()
```

# 5) Hyperparameter Tuning

```python
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import GridSearchCV
import time
```

```python
# Param grids
param_grid = {
    'hidden_layer_sizes': [(50,50,50), (50,100,50), (100,)],
    'activation': ['tanh', 'relu'],
    'solver': ['sgd', 'adam'],
    'alpha': [0.0001, 0.05],
    'learning_rate': ['constant','adaptive'],
}

# Grid Search for tuning 
mlp = MLPClassifier()
grid_search = GridSearchCV(estimator=mlp, param_grid=param_grid, n_jobs=-1, cv=2, scoring='accuracy')

start = time.time()
grid_search.fit(X_train, y_train)
end = time.time()
```

```python
print("Tuning time of model is: ", end-start, "(s)")
```

```python
# Best Values
best_params = grid_search.best_params_
best_estimator = grid_search.best_estimator_

# Print the best parameters
print("Best Parameters found by GridSearchCV:")
print(best_params)
```

```python
# Evaluation of best estimator
y_pred_best_mlp = best_estimator.predict(X_test)
y_proba_best_mlp = best_estimator.predict_proba(X_test)
comprehensive_evaluation(y_test, y_pred_best_mlp, class_names, y_proba_best_mlp)
```
