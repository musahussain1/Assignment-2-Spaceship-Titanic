## 3.1) Univariate Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pie chart for distribution of feature
def plot_pie_chart(df, feature, figsize=(6, 6), color_map='pastel', pctdistance=0.7, explode_factor=0.02):
    # Count the occurrences of each class in the feature
    target_counts = df[feature].value_counts()
    
     # Creating a pie chart
    plt.figure(figsize=figsize)
    colors = sns.color_palette(color_map)[0:len(target_counts)]
    explode = [explode_factor] * len(target_counts)  # Explode each slice by the specified factor
    plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', colors=colors, startangle=140, explode=explode, pctdistance=pctdistance)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# Pie plot of Transported
plot_pie_chart(training_data, 'Transported')
```

```python
# Pie plot of Transported
plot_pie_chart(training_data, 'CryoSleep', color_map='muted')
```

```python
# Function to analyze categorical variable
def plot_bar_chart(df, feature, all=True):
    # Counting the occurrences of each category
    feature_counts = df[feature].value_counts()

    if all:
        pass
    else:
        feature_counts = feature_counts.head(10)
        
    feature_percentages = (feature_counts / len(df)) * 100

    print(f"Counts for {feature}:")
    print(feature_counts, "\n")
    print(f"Percentages for {feature}:")
    print(feature_percentages, "\n")
    
    # Creating a bar plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x=feature_counts.index, y=feature_counts.values, palette='pastel')
    
    # Adding percentages on top of the bars
    for index, value in enumerate(feature_counts.values):
        plt.text(index, value, f'{value} ({feature_percentages.values[index]:.1f}%)', ha='center', va='bottom')
    
    plt.title(f'Bar Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

## 3.2) Bivariate Analysis

```python
def transportation_rate_by_category(df, categorical_feature, figsize=(10, 6)):
    target_feature='Transported'
    
    # Calculate transportation rate for each category
    category_counts = df.groupby(categorical_feature)[target_feature].mean() * 100
    
    # Print transportation rates
    print(f"Transportation rates for {categorical_feature}:")
    print(category_counts, "\n")
    
    # Create a bar plot
    plt.figure(figsize=figsize)
    sns.barplot(x=category_counts.index, y=category_counts.values)
    
    # Add percentages on top of the bars
    for index, value in enumerate(category_counts.values):
        plt.text(index, value, f'{value:.1f}%', ha='center', va='bottom')
    
    plt.title(f'Transportation Rate by {categorical_feature}')
    plt.xlabel(categorical_feature)
    plt.ylabel('Transportation Rate (%)')
    plt.show()
```

```python
# Analyzing transportation rate by HomePlanet
transportation_rate_by_category(training_data, 'HomePlanet')
```

```python
# Analyzing transportation rate by CryoSleep
transportation_rate_by_category(training_data, 'CryoSleep', (5, 4))
```

```python
# Analyzing transportation rate by Destination
transportation_rate_by_category(training_data, 'Destination')
```

```python
# Analyzing transportation rate by VIP
transportation_rate_by_category(training_data, 'VIP', (6, 4))
```

```python
def transportation_rate_by_numerical_feature(df, numerical_feature, bins=10):
    target_feature='Transported'
    
    # Create bins for the numerical feature
    df['binned'] = pd.cut(df[numerical_feature], bins=bins)
    
    # Calculate transportation rate for each bin
    bin_transportation_rate = df.groupby('binned')[target_feature].mean() * 100
    
    # Print transportation rates
    print(f"Transportation rates for {numerical_feature} (binned):")
    print(bin_transportation_rate, "\n")
    
    # Create a bar plot
    plt.figure(figsize=(12, 6))
    sns.barplot(x=bin_transportation_rate.index.astype(str), y=bin_transportation_rate.values, palette='pastel')
    
    # Add percentages on top of the bars
    for index, value in enumerate(bin_transportation_rate.values):
        plt.text(index, value, f'{value:.1f}%', ha='center', va='bottom')
    
    plt.title(f'Transportation Rate by {numerical_feature} (binned)')
    plt.xlabel(f'{numerical_feature} (binned)')
    plt.ylabel('Transportation Rate (%)')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Transportation rate by RoomService
transportation_rate_by_numerical_feature(training_data, 'RoomService', 5)
```

```python
# Transportation rate by Age
transportation_rate_by_numerical_feature(training_data, 'Age', 5)
```

```python
# Transportation rate by FoodCourt
transportation_rate_by_numerical_feature(training_data, 'FoodCourt', 5)

# Transportation rate by ShoppingMall
transportation_rate_by_numerical_feature(training_data, 'ShoppingMall', 5)

# Transportation rate by Spa
transportation_rate_by_numerical_feature(training_data, 'Spa', 5)

# Transportation rate by VRDeck
transportation_rate_by_numerical_feature(training_data, 'VRDeck', 5)
```

I fill missing numerical values with the median and categorical with the most frequent value

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.impute import SimpleImputer

# Columns by type
numeric_features = ['RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck', 'Age']
categorical_features = ['HomePlanet', 'CryoSleep', 'Cabin', 'Destination', 'VIP']
```

```python
# Pipeline for imputing null in numeric and Feature Scaling
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

# Encoding Categorical Features and Scaling Numerical Features
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', Pipeline(steps=[
            ('impute', SimpleImputer(strategy='most_frequent')),
            ('onehot', OneHotEncoder(handle_unknown='ignore'))
        ]), categorical_features)
    ])
```

```python
# Preprocessing the data
X = training_data.drop(columns=['Transported', 'PassengerId', 'Name'])
y = training_data['Transported']

# Applying transformations to the training data
X_preprocessed = preprocessor.fit_transform(X)
```

```python
# Converting the preprocessed data back to a DataFrame
preprocessed_columns = numeric_features + list(preprocessor.named_transformers_['cat']['onehot'].get_feature_names_out(categorical_features))
X_preprocessed_df = pd.DataFrame(X_preprocessed.toarray(), columns=preprocessed_columns)
X_preprocessed_df.head()
```
