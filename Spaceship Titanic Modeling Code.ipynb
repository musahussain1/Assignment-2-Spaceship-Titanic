# 1) Reading Data

```python
import pandas as pd

# Reading data files
training_data = pd.read_csv("https://raw.githubusercontent.com/musahussain1/Assignment-2-Spaceship-Titanic/main/spaceship_titanic_dataset/train.csv")
testing_data = pd.read_csv("https://raw.githubusercontent.com/musahussain1/Assignment-2-Spaceship-Titanic/main/spaceship_titanic_dataset/test.csv")
sample_data = pd.read_csv("https://raw.githubusercontent.com/musahussain1/Assignment-2-Spaceship-Titanic/main/spaceship_titanic_dataset/sample_submission.csv")
```

# 2) Exploring Data

```python
def explore_data(df, df_name="DataFrame"):
    print(f"Exploration of {df_name}:")
    
    # Display the number of rows and columns
    print(f"Number of rows: {df.shape[0]}")
    print(f"Number of columns: {df.shape[1]}\n")
    
    # Display the first few rows of the dataframe
    print("First few rows of the dataframe:")
    display(df.head())
    print()
    
    # Display basic information about the dataframe
    print("Basic Information about the dataframe:")
    display(df.info())
    print()
    
    # Display summary statistics of the dataframe
    print("Summary statistics of the dataframe:")
    display(df.describe(include='all'))
    print()
    
    # Display the number of missing values in each column
    print("Missing values in each column:")
    display(df.isnull().sum())
    print()
    
    # Display the number of unique values in each column
    print("Unique values in each column:")
    display(df.nunique())
```

```python
# Exploration on training data
explore_data(training_data, "Training Data")
```

```python
# Exploration on testing data
explore_data(testing_data, "Testing Data")
```

```python
# Exploration on sample data
explore_data(sample_data, "Sample Submission Data")
```

# 3) Data Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Pie chart for distribution of feature
def plot_pie_chart(df, feature, figsize=(6, 6), color_map='pastel'):
    # Count the occurrences of each class in the feature
    target_counts = df[feature].value_counts()
    
    # Creating a pie chart
    plt.figure(figsize=figsize)
    colors = sns.color_palette(color_map)[0:len(target_counts)]
    plt.pie(target_counts, labels=target_counts.index, autopct='%1.1f%%', colors=colors, startangle=140)
    plt.title(f'Distribution of {feature}')
    plt.show()
```

```python
# Pie plot of Transported
plot_pie_chart(training_data, 'Transported')
```

```python
# Pie plot of Transported
plot_pie_chart(training_data, 'CryoSleep', color_map='muted')
```

```python
# Function to analyze categorical variable
def plot_bar_chart(df, feature, all=True):
    # Counting the occurrences of each category
    feature_counts = df[feature].value_counts()

    if all:
        pass
    else:
        feature_counts = feature_counts.head(10)
        
    feature_percentages = (feature_counts / len(df)) * 100

    print(f"Counts for {feature}:")
    print(feature_counts, "\n")
    print(f"Percentages for {feature}:")
    print(feature_percentages, "\n")
    
    # Creating a bar plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x=feature_counts.index, y=feature_counts.values, palette='pastel')
    
    # Adding percentages on top of the bars
    for index, value in enumerate(feature_counts.values):
        plt.text(index, value, f'{value} ({feature_percentages.values[index]:.1f}%)', ha='center', va='bottom')
    
    plt.title(f'Bar Plot of {feature}')
    plt.xlabel(feature)
    plt.ylabel('Count')
    plt.show()
```

```python
plot_bar_chart(training_data, 'HomePlanet')
```

```python
plot_bar_chart(training_data, 'Cabin', all=False)
```
